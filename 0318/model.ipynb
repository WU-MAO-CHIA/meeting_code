```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
(x_train ,y_train), (x_valid, y_valid) = mnist.load_data()
```

```python
#flatten 28*28 = 784
x_train = x_train.reshape(60000, 784)
x_valid = x_valid.reshape(10000, 784)
```

```python
x_train[0]
```

```python
#normailzation
x_train = x_train / 255
x_valid = x_valid / 255
```

```python
x_train.max()
```

```python
x_train[0]
```

```python
x_train.dtype
```

```python
y_train[0:9]
```

```python
# we do one hot encoding
# 建立一個 10 維的向量，因為數字產生的距離差是沒有意義的
# true answer 是第一類，但預測出來是第二類跟預測出來是第三類的loss應該相同

num_categories = 10

y_train = tf.keras.utils.to_categorical(y_train, num_categories)
y_valid = tf.keras.utils.to_categorical(y_valid, num_categories)
```

```python
y_train[0:9]
```

```python
model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(512, activation= 'relu', input_shape= (784,)),
        tf.keras.layers.Dense(512, activation= 'relu'),
        tf.keras.layers.Dense(10, activation= 'softmax')
])
#做分類問題最後都要進 softmax 把值壓到 0~1 之間才能當作機率
```

```python
model = keras.Sequential()
model.add(layers.Dense(512, activation= 'relu', input_shape= (784,)))
model.add(layers.Dense(256, activation= 'relu'))
model.add(layers.Dense(128, activation= 'relu'))
model.add(layers.Dense(10, activation= 'softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer= 'adam' ,loss= 'categorical_crossentropy', metrics= ['accuracy'])
```

```python
history = model.fit(
    x_train, y_train, epochs= 10, verbose= 1, validation_data= (x_valid, y_valid)
)
```

```python
model.evaluate(x_valid, y_valid, verbose= 1)
```

```python
model.save('mnist')
```
